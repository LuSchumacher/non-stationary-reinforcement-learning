```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FONT_SIZE_1 = 22
FONT_SIZE_2 = 18
FONT_SIZE_3 = 16

import matplotlib
matplotlib.rcParams['font.serif'] = "Palatino"
matplotlib.rcParams['font.family'] = "serif"
```

```python
time = np.arange(1, 401)
y1 = np.repeat([1.75], 400)
y2 = np.exp(np.linspace(1, 0, 400))
param_values = [y1, y2]
titles = ["Static", "Deterministic function"]
```

```python
fig, axarr = plt.subplots(1, 2, figsize=(10, 3), sharey=True)
for i, ax in enumerate(axarr.flat):
    ax.plot(
        time,
        param_values[i],
        color='maroon'
    )

    ax.fill_between(
        time,
        param_values[i] - 0.2,
        param_values[i] + 0.2,
        color='maroon', alpha=0.5, linewidth=0.0
    )
    ax.set_title(f'{titles[i]}', fontsize=FONT_SIZE_1)
    ax.tick_params(axis='both', which='major', labelsize=FONT_SIZE_3)
    ax.set_xticks([])
    ax.set_yticks([])
    if i == 0:
        ax.set_ylabel("Parameter value", fontsize=FONT_SIZE_2, labelpad=10)
    ax.set_xlabel("Trial", fontsize=FONT_SIZE_2, labelpad=15)

sns.despine()
fig.tight_layout()
fig.savefig("../plots/traditional_approaches.pdf", dpi=300, bbox_inches="tight")
```
